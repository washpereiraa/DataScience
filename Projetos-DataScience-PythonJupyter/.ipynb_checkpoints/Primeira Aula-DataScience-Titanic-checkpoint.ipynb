```python
#transformei o codigo acima numa funcao
#nesta funcao recebe uma linha da lista por vez mais o indice da linha
def tratarNome(linha,i):
    import re
    match = re.compile(r'"(.*)(,)(\s.*)"')
    name = match.sub(r'"\3  \1"', linhas[i])
    return name
```

```python
 linhaTratada = tratarNome(linhas[2],2)
```

```python
linhaTratada
```

```python
#funcao para tratar todas as linhas e fazer um novo texto
texto = [] #declaracao de lista vazia
texto2 = []
def txtConcatenado (text,i):
    texto.clear()
    linhaTratada = tratarNome(text,i)
    texto.append(linhaTratada)
    texto2.append(linhaTratada)
    #print(linhaTratada)
    return texto
```

```python
#chama a funcao dentro de um lao para percorrer todas as linhas do texto e enviar como parametro
texto.clear() #retorno de uma linha tratada
texto2.clear() #o texto todo tratado
tamanhotexto = len(linhas)
for i in range(1, tamanhotexto):
    print(txtConcatenado(linhas[i],i))
```

```python
#criar um cicionario a partir da lista (dados do banco)
titanicDados = {} #declarao de variavel de dados
for c in colunas: #variavel colunas ja esta declarada em algum ponto desse documento
    titanicDados[c] = [] #cada coluna rece uma lista vazia    
    
titanicDados
    
```

```python
#segunda forma (mais facil) de fazer a msm funcao criada acima
def tratarNome2(linha):
    import re
    match = re.compile(r'"(.*)(,)(\s.*)"')
    name = match.sub(r'"\3  \1"', linha)
    return name
```

```python
#uma segunda forma (mais facil) de fazer o for percorrer uma lista
for dado in linhas[1:]:
    dadoTratado = tratarNome2(dado)
    print(dadoTratado)
```

```python
#tratando o dado e separando em lista
for dado in linhas[1:]:
    dadoTratado = tratarNome2(dado)
    dadoTratadoSeparado = dadoTratado.split(',')
    print(dadoTratadoSeparado)
```

```python
#fazendo uma lista de lista - terceira parte do cotigo ja feito acima
ListaDeListas = []
for dado in linhas[1:]:
    dadoTratado = tratarNome2(dado)
    dadoTratadoSeparado = dadoTratado.split(',')
    ListaDeListas.append(dadoTratadoSeparado)
```

```python
print(colunas[1], colunas[3])
print(ListaDeListas[1][1],ListaDeListas[1][3])
```

```python
for passageiro in ListaDeListas:
    print(colunas[3], passageiro[3],colunas[5],passageiro[5],colunas[4], passageiro[4])
```

```python
for passageiro in ListaDeListas:
    print(passageiro)
    for indice, dado in enumerate (passageiro):
        print('INDICE:', indice, "\n     DADOS:", dado)
```

```python
for passageiro in ListaDeListas:
    for indice, dado in enumerate(passageiro):
        coluna = colunas[indice]
        titanicDados[coluna].append(dado)
```

```python
titanicDados
```

```python
for idade in titanicDados['Age']:
    print(idade)
```

```python
for indice, idade in enumerate(titanicDados['Age']):
    if titanicDados['Age'][indice] != '':
        titanicDados['Age'][indice] = float(idade)
    else:
        titanicDados['Age'][indice] = -1
```

```python
listaIdades = []

for idade in titanicDados['Age']:
    if idade != -1:
        listaIdades.append(idade)
```
