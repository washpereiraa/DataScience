```python
import pandas as panda
import numpy

colunas = {'star_rating':str, 'title':str, 'content_rating':str, 'genre':str, 'duration':float, 'actors_list':str}
analiseFilmes = panda.read_csv("C:\GIT\DataScience\Datasets\movies.csv", header=0, names=colunas)
analiseFilmes.head(5)
```

```python
analiseFilmes.dtypes
```

```python
#EXPLORANDO SERIES NAO NUMERICAS

#a gente pode utilizar o describe por coluna
analiseFilmes.genre.describe()
```

```python
#Value_counts - funciona de forma parecida com o groupby

analiseFilmes.genre.value_counts()
```

```python
#Value_counts - normalize - exibindo por porcetagem

analiseFilmes.genre.value_counts(normalize=True)
```

```python
#Unique - dados que nao se repetem

analiseFilmes.genre.unique()
```
