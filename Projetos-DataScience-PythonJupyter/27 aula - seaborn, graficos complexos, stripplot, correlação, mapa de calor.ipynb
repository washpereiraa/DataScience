```python
import seaborn
import matplotlib.pyplot as matplot
%matplotlib inline

#analisando o arquivo TIPS - Gorgeta
gorgeta = seaborn.load_dataset('tips')
gorgeta.head(5)
```

```python
#STRIPPLOT

seaborn.stripplot(x='day', y='tip', hue='time', data=gorgeta)
```

```python
#SWARPLOT
seaborn.swarmplot(x='day', y='tip', data=gorgeta)
seaborn.violinplot(x='day', y='tip', color='black', data=gorgeta)
```

```python
#VISUALIZANDO CORRELAO/MAPA DE COLOR


correlacaoGorgeta = gorgeta.corr() #dados dos graticos em numeros
correlacaoGorgeta 
```

```python
#HEATMAP MAPA DE COR
#Quanto maior a correlo, mais igual so as cores

seaborn.heatmap(correlacaoGorgeta, annot=True)
```

```python
#analises do dataset flights por mapa de cores

voos = seaborn.load_dataset('flights')
voos = voos.pivot_table(values='passengers', index='month', columns='year')
voos.head(5)
```

```python
seaborn.heatmap(voos, cmap='OrRd')
```

```python
#clustermap
#agrupa indices com mais ocorrencias 
seaborn.clustermap(voos, cmap='OrRd')
```

```python
#LMPLOT
#faz previso das futuras ocorrencias
seaborn.lmplot(x='total_bill', y='tip', col='time', hue='time', data=gorgeta)
```
