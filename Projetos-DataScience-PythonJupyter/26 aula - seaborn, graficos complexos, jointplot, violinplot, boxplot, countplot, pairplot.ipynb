```python
import seaborn
import matplotlib.pyplot as matplot
%matplotlib inline

#analisando o arquivo TIPS - Gorgeta
gorgeta = seaborn.load_dataset('tips')
gorgeta.head(5)
```

```python
# GRAFICOS LINEARES
# graficos linerares: conforme o y aumente, x tbm aumenta e virse e versa.

#Quanto maior o valor da conta, maior o valor da gorgeta? Vamos ver

seaborn.jointplot(x='total_bill', y='tip', data=gorgeta)
#sim quanto maior o valor da conta, maior a gorgete
```

```python
#quanto mais lugares a mesa, maior a gorgeta?

seaborn.jointplot(x='size', y='tip', data=gorgeta)
#n√£o, meses com menos lugares deram mais gorgeta
```

```python
#Analisando arquivo Titanic

titanic = seaborn.load_dataset('titanic')
titanic.head(5)
```

```python
#Quanto maior a idade, maior o valor da passagem?

seaborn.jointplot(x='age', y='fare', data=titanic)
#n√£o, a maioria pagou entre 0 a 100
```

```python
#PAIRPLOT
#pairplot gera varios graficos para analise
#na diagonal s√£o graficos que cruzam os mesmos dados - sem uso

#hue diferencia valores dos dados por cores, a coluna time √© composta por 'jantar/almo√ßo'
seaborn.pairplot(gorgeta, hue="sex", palette="dark", kind="scatter" )

#paleta de cores
#https://chrisalbon.com/python/data_visualization/seaborn_color_palettes/
```

```python
#COUNTPLOT
#contar dados categoricos
seaborn.countplot(x='time', data=gorgeta)
```

```python
#BOXPLOT
#separa dado por quartis

seaborn.boxplot(x='day', y='tip', data=gorgeta)
```

```python
#VIOLINPLOT

seaborn.violinplot(x='day', y='tip', hue='time', split=True, data=gorgeta)
```
